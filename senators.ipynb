```python
import requests
import lxml.html
import pandas as pd
```

```python
url = 'https://www.senate.gov/senators/index.htm'
xpath = '//tr'
```

```python
response = requests.get(url)
open('response.html','w').write(response.text)
df_senate = pd.read_html('response.html')[0].sort_values('State')
```

#### Show all Senators by state

```python
# df_pop = pd.read_csv('state_populations.csv')
df_pop = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTjGMuG3E7qkX7gyGwOHiRhhX4cxsPXJTvCSohCq1QqnHZ0t7ImauI1rwPc9OO0pIXfvflJGv12ZzeG/pub?gid=2136116875&single=true&output=csv')
df_senate2 = df_senate.merge(df_pop[['State','Pop']],on='State',how='inner')
pd.set_option('display.max_rows', len(df_senate2)+1)
df_senate2
```

#### Show Republican controled small population states (less than 1 million citizens)

```python
df_senate3 = df_senate2[df_senate2.Party=='Republican']
df_small_pop_republican_states = df_senate3[df_senate3.Pop<=1000000]
df_small_pop_republican_states = df_small_pop_republican_states.rename(
    columns={"Senator's Name":'Senator'}
)
df_small_pop_republican_states = df_small_pop_republican_states.reset_index()
df_small_pop_republican_states = df_small_pop_republican_states[
    ["Senator","State",'Pop']
]
df_small_pop_republican_states = df_small_pop_republican_states.sort_values(
    ["State","Senator"]
)
df_small_pop_republican_states = df_small_pop_republican_states.groupby(
    ['State','Pop']).apply(
    lambda df:pd.Series({'Senators':df.Senator.values})
).reset_index()

df_small_pop_republican_states
```

#### Show total combined population of these states

```python
total_pop = df_small_pop_republican_states.Pop.sum()
total_pop = "{:,}".format(total_pop)
print(
    f'''
    The total combined population of states with less than a million people, 
    and 2 Republican Senators: {total_pop}
    '''
)
```
